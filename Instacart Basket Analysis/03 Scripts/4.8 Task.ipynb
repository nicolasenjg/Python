### Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(20)
```

### The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified.

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
# whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# The team now wants to target different types of spenders in their marketing campaigns. we have to split into 2 
#first getting the average based on the user_id and apply transform
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
# Determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s 
# ordering behavior according to the median in the “days_since_prior_order” column. 
```

```python
ords_prods_merge['Frecuency'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['Frecuency'] > 20, 'ord_freq_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['Frecuency'] > 10) & (ords_prods_merge['Frecuency'] <= 20), 'ord_freq_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['Frecuency'] <= 10, 'ord_freq_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['ord_freq_flag'].value_counts(dropna=False)
```
